### **Variabel AGE**

Variabel pertama yang akan digunakan untuk mengidentifikasi distribusi terbaik yang digunakan yaitu AGE.

Terlebih dahulu, kita perlu mem-ploting data menggunakan histogram atau displot untuk mendapatkan gambaran awal tentang distribusi data tersebut.

disini digunakan `distplot()` dari library seaborn

````python
sns.displot(df["AGE"], kde=True, color='teal', bins=20)
````

Berdasarkan output displot diatas, dapat dilihat bahwa pasien diabetes didominasi oleh pasien dengan rentang umur 50-65 tahun. Hal ini menyebabkan kurva menceng ke kiri atau memiliki skewness negatif.

Sebelum kita menggunakan Fitter, kita perlu mengonversi data atau variabel yang akan diguankan menjadi NumPy array. Salah satu caranya yaitu menggunakan atribut `.values`

````python
Age=df['AGE'].values
````

Pada dasarnya, untuk menentukan distribusi yang sesuai, dapat langsung menggunakan code `Fitter(Value)`, seperti berikut:

````python
fit=Fitter(Age)
fit.fit()
fit.summary()
````

`summary()` akan menampilkan 5 distribusi yang berbeda dan nilai sumsquare_error, kriteria informasi Akaike (aic), dan kriteria informasi Bayesian (bic) pada masing-masing distribusi. Secara default, function `summary` mengurutkan lima distribusi terbaik berdasarkan nilai `sumsquare_error`. Selain itu, fungsi ini juga memberikan ilustrasi dari berbagai distribusi dengan histogram.

Distribuasi terbaik dapat ditentukan berdasarkan nilai `sumsquare_error`. Semakin kecil nilai tersebut, maka semakin sesuai distribusi dengan data.

Berdasarkan nilai `sumsquare_error`, dari keseluruhan distributions (110 distribusi) distribusi terbaik untuk data AGE adalah distribusi triang.

Fitter pada dasarnya menggunakan library Scipy yang mendukung 110 distribusi. Fitter akan memindai semuanya, dan memberikan ringkasan distribusi terbaik berdasarkan sum square errors.

Namun, hal ini memungkinkan memakan waktu lebih lama karena akan mencoba semua distribusi apalagi jika ukuran sampel yang digunakan besar. Jadi, disarankan untuk memplot histogram terlebih dahulu dan mendapatkan gambaran umum tentang jenis distribusi yang mungkin sesuai dengan data dan membatasi Fitter dalam menentukan distribusi berdasarkan daftar distribusi yang dipilih. Hal ini akan lebih menghemat waktu.

Disini, distribusi dibatasi pada 5 distribusi yang dipilih yaitu gamma, lognorm, beta, burr, dan normal.

````python
# Menentukan distribution terbaik untuk data variabel AGE
fit_AGE = Fitter(Age,
           distributions=['gamma',
                          'lognorm',
                          "beta",
                          "burr",
                          "norm"])
fit_AGE.fit()
fit_AGE.summary()
````

Berdasarkan nilai `sumsquare_error`, dari kelima distribusi yang dipilih, distribusi terbaik untuk data AGE adalah **burr distribution**.
